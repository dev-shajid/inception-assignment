```python
'''
1. Email Validator (String Filtering + Loops)
ðŸ’¼ Real-world use: Used in signup forms or data cleaning tasks.
Task:
Check if a given email is valid â€” must contain â€œ@â€ and end with â€œ.comâ€ or â€œ.orgâ€.
Example:
Input: hello@inceptionbd.com
Output: Valid Email
'''
email = input("Enter email: ").split('@')

first = email[0]
last = email[-1]

ends_with_valid_domain = (last.endswith(".com") or last.endswith(".org"))

if len(first)>0 and len(last)>3 and ends_with_valid_domain:
    print("Valid Email")
else:
    print("Invalid Email")
```

```python
'''
2. Count Word Frequency in Customer Reviews
ðŸ’¼ Real-world use: Analyzing reviews or comments in text analytics.
Task:
Given a review text, count how many times each word appears (ignore case).
Example:
Input: "This product is good and this service is excellent"
Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}
'''
text = input("Enter review: ").lower()
words = text.split()
word_count = []

for word in words:
    found = False
    for pair in word_count:
        if pair[0] == word:
            pair[1] += 1
            found = True
            break
    if not found:
        word_count.append([word, 1])
        
print("{", end="")
for i in range(len(word_count)):
    print(f"'{word_count[i][0]}': {word_count[i][1]}", end="")
    if i != len(word_count) - 1:
        print(", ", end="")
print("}")
```

```python
'''
3. Password Strength Checker
ðŸ’¼ Real-world use: Used in user authentication systems.
Task:
Check if a password is strong â€” must have at least one uppercase, one lowercase, one digit, and one
special character.
Example:
Input: "Hello@123"
Output: Strong Password
'''
password = input("Enter password: ")

has_upper = False
has_lower = False
has_digit = False
has_special = False

for ch in password:
    if ch.isupper():
        has_upper = True
    elif ch.islower():
        has_lower = True
    elif ch.isdigit():
        has_digit = True
    else:
        has_special = True

if has_upper and has_lower and has_digit and has_special:
    print("Strong Password")
else:
    print("Weak Password")
```

```python
'''
4. Username Generator
ðŸ’¼ Real-world use: Automatically create usernames for users.
Task:
Take userâ€™s full name and generate a short username.
Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).
Example:
Input: "Mamun Malitha"
Output: mamitha57
'''
import random

name = input("Enter full name: ").split()
first = name[0]
last = name[-1]

username = first[:3].lower() + last[-3:].lower()

random_number = ""
random_number += str(random.randint(10, 90))

username += random_number
print(username)
```

```python
'''
5. SMS Word Counter
ðŸ’¼ Real-world use: Used in messaging apps or social media post limit check.
Task:
Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).
Example:
Input: "This is a demo message."
Output: Words = 5, Characters = 23, Status = Within Limit
'''
msg = input("Enter message: ").strip()

words = msg.split()
word_count = len(words)
char_count = len(msg)

if char_count > 160:
    status = "Exceeded Limit"
else:
    status = "Within Limit"

print(f"Words = {word_count}, Characters = {char_count}, Status = {status}")
```

```python
'''
6. Remove Duplicates from Contact List
ðŸ’¼ Real-world use: Data cleaning in CRM or user databases.
Task:
Given a list of names (comma separated), remove duplicates.
Example:
Input: "Rahim, Karim, Rahim, Sultana"
Output: "Rahim, Karim, Sultana"
'''
contacts = input("Enter names (comma separated): ")
names = contacts.split(",")

unique = []
for n in names:
    n = n.strip()
    if n not in unique:
        unique.append(n)

print(", ".join(unique))
```

```python
'''
7. Detect Suspicious Words in Text (Moderation)
ðŸ’¼ Real-world use: Used in chat moderation or email filtering.
Task:
If any â€œbannedâ€ words like ['spam', 'scam', 'fake'] are found in a text, print â€œWarningâ€.
Example:
Input: "This is a scam offer"
Output: Warning: Suspicious content detected
'''
text = input("Enter text: ").lower()
banned = ['spam', 'scam', 'fake']

for w in banned:
    if w in text:
        print("Warning: Suspicious content detected")
        break
else:
    print("No suspicious content")
```

```python
'''
8. Invoice Code Formatter
ðŸ’¼ Real-world use: Formatting IDs or codes in finance or ERP systems.
Task:
Given an invoice number like 23, generate code in the format INV00023.
Example:
Input: 23
Output: INV00023
'''
num = int(input("Enter invoice number: "))
code = "INV" + str(num).zfill(5)
print(code)
```

```python
'''
9. Count Hashtags and Mentions (Social Media Analytics)
ðŸ’¼ Real-world use: Used in analyzing social media posts.
Task:
Count how many hashtags (#) and mentions (@) appear in a post.
Example:
Input: "Check out #AIWorkshop and follow @inceptionbd"
Output: Hashtags = 1, Mentions = 1
'''
text = input("Enter post: ").strip()
words = text.split()

hashtags = 0
mentions = 0

for w in words:
    if len(w) > 1:
        if w.startswith('#'):
            hashtags += 1
        elif w.startswith('@'):
            mentions += 1

print("Hashtags =", hashtags, ", Mentions =", mentions)
```

```python
'''
10. Word Reversal for Data Encryption
ðŸ’¼ Real-world use: Simple text obfuscation or playful encoding.
Task:
Reverse each word in a sentence while keeping word order the same.
Example:
Input: "AI is fun"
Output: "IA si nuf"
'''
sentence = input("Enter sentence: ").strip()
words = sentence.split()
reversed_words = []

for w in words:
    reversed_words.append(w[::-1])

print(" ".join(reversed_words))
```
