## 1: Checks the temperature and prints a corresponding message.

```python
temp = int(input("Enter the current temperature in Celsius: "))

if temp < 10:
    print("It's too cold!")
elif temp <= 25:
    print("It's cool outside")
else:
    print("It's hot outside!")
```

## 2: ATM Withdrawal System

```python
balance = float(input("Enter your current account balance: BDT "))
withdrawal = float(input("Enter the amount you want to withdraw: BDT "))

if withdrawal > balance:
    print("Insufficient Balance")
else:
    remaining_balance = balance - withdrawal
    print("Transaction Successful!")
    print(f"Remaining Balance: BDT {remaining_balance:.2f}")
```

## 3: Calculates a student's grade based on marks (0-100)

```python
marks = int(input("Enter student's marks (0-100): "))

if not (0 <= marks <= 100):
    print("Marks must be between 0 and 100.")
    return

if marks >= 90:
    print("Grade: A+")
elif marks >= 80: 
    print("Grade: A")
elif marks >= 70: 
    print("Grade: B")
elif marks >= 60: 
    print("Grade: C ")
else: # Below 60
    print("Grade: Fail")
```

## 7: Prints action based on the input traffic light color.

```python
color = input("Enter the traffic light color (red, yellow, or green): ").lower()

if color == "red":
    print("Stop")
elif color == "yellow":
    print("Ready to go")
elif color == "green":
    print("Go")
else:
    print("Invalid color. Please enter red, yellow, or green.")
```

## 8: Determines if a year is a leap year using the standard rules.

```python
year = int(input("Enter a year: "))

if year % 400 == 0:
    print(f"{year} is a Leap Year!")
elif (year % 4 == 0) and (year % 100 != 0):
    print(f"{year} is a Leap Year!")
else:
    print(f"{year} is Not a Leap Year")
```

## 10. Electricity Bill Calculator

```python
units = int(input("Enter number of units consumed: "))
total_bill = 0

if units <= 100:
    total_bill = units * 5
elif units <= 200:
    bill_for_100 = 100 * 5 
    remaining_units = units - 100
    bill_for_rest = remaining_units * 7
    total_bill = bill_for_100 + bill_for_rest
else:
    bill_for_100 = 100 * 5
    bill_for_next_100 = 100 * 7
    remaining_units = units - 200
    bill_for_rest = remaining_units * 10
    total_bill = bill_for_100 + bill_for_next_100 + bill_for_rest

print(f"Total Payable Amount: BDT {total_bill:.2f}")
```
